```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

#Loading the MNIST dataset 
data, labels = load_digits(return_X_y=True) 
```

```python
n_components=3
pca = PCA(n_components=3)
data_pca = pca.fit_transform(data)
```

```python
n_components=3
perplexity = 15
tsne = TSNE(n_components=3, perplexity = 15)
data_tsne = tsne.fit_transform(data_pca)
```

```python
plt.figure(figsize= (12,8))

for x in range(10):
    flag = (labels == x)
    plt.scatter(data_pca[flag,0], data_pca[flag, 1],label= str(x), cmap = 'tab10')
   
plt.xlabel("Principal Component of x axis")  
plt.ylabel("Principal Component of y axis")  
plt.title("PCA Visualization of MNIST Digits")
plt.legend(title = "Digits")

cbar=plt.colorbar(ticks=range(10))
cbar.set_label('Digits', rotation=90)
plt.show()
```

```python
plt.figure(figsize= (12,8))

for x in range(10):
    flag = (labels == x)
    plt.scatter(data_tsne[flag,0], data_tsne[flag, 1],label= str(x), cmap = plt.get_cmap('tab10'))
    
plt.xlabel("t-SNE Dimension 1")  
plt.ylabel("t-SNE Dimension 2")  
plt.title("t-SNE of MNIST Digits")
plt.legend(title = "Digits")

cbar=plt.colorbar(ticks=range(10))
cbar.set_label('Digits', rotation=90)
plt.show()
```
