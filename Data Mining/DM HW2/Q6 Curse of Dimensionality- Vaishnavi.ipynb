```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
import math
```

```python
dimensions = list(range(2, 51))
distance= []

for s in dimensions:
    data = [np.random.randint(-100, 100, s) for _ in range(10000)]
    datapoint= np.random.randint(-100, 100, s)
    euclidean = [np.linalg.norm(p - datapoint) for p in data]
    max_diff = max(euclidean)
    min_diff = min(euclidean)
    avg_distance = math.log(max_diff - min_diff) / min_diff
    distance.append(avg_distance)

```

```python
plt.figure(figsize=(12, 8))
plt.plot(dimensions, distance, marker='o', linestyle='-')
plt.title('Graph for Curse of Dimensionality')
plt.xlabel('Dimensionality of Graph')
plt.ylabel('log((MAX_DIST - MIN_DIST) / MIN_DIST)')
plt.grid(True)
plt.show()
```
